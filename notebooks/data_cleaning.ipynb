```python
import pandas as pd, numpy as np, sklearn

my_test = pd.read_csv("./nooz/scripts/2016-09-27.csv", header=None)
print my_test
```

```python
import datetime, dateutil, string

def process(line):
    naive_split = line.split(',')
    try:
        timestamp = dateutil.parser.parse(naive_split[0])
    except:
        print "failed on", line
        raise Exception
    title = naive_split[1]
    url = naive_split[-1]
    summary = string.join(naive_split[2:-1])
    
    ## Server saves GMT by default but we're five hours behind
    
    time_adjusted = timestamp - datetime.timedelta(hours=5)
    
    new_line = str(time_adjusted) + '|' + title + '|' + summary + '|' + url
    
    return new_line

my_dates = ["2016-09-26", "2016-09-27", "2016-09-29", "2016-09-30", "2016-10-03"]

for date in my_dates:
    this_file = open("./nooz/scripts/"+date+".csv", "r")
    new_file = open("./nooz/scripts/"+date+"_fixed.csv", "w")
    for line in this_file.readlines():
        try:
            new_file.write(process(line)+"\n")
        except:
            print date
    this_file.close()
    new_file.close()
```

```python
frame_list=[]

for date in my_dates:
    frame_list.append(pd.read_csv("./nooz/scripts/"+date+"_fixed.csv", sep="|", header=None))

big_frame = pd.concat(frame_list).drop_duplicates(subset=[1], keep="first")
print big_frame
```
