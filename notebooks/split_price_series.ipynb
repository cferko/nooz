```python
import pandas as pd, numpy as np, glob
```

```python
my_files = glob.glob("/home/lcarter/energy/*.csv")
for f in [f for f in my_files if "2016" in f]:
    this_df = pd.read_csv(f)
    symbols = np.unique(this_df["StockSymbol"].values)
    
    this_date = f.split("/")[-1].split('.')[0]
    
    for symbol in symbols:
        this_split = this_df[this_df["StockSymbol"]==symbol]
        this_split.to_csv("/home/cferko/split_prices/"+symbol+"_"+this_date+".csv")
```
